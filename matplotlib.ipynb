```python
#4
import matplotlib.pyplot as plt
import csv

dates = []
closing_prices = []

file = open("fdata.csv")
reader = csv.reader(file)
next(reader)

for row in reader:
    dates.append(row[0])
    closing_prices.append(float(row[4]))

file.close()

plt.plot(dates, closing_prices, marker='o')
plt.xlabel("Data")
plt.ylabel("Closing Price")
plt.title("Alphabet Inc. Closing Prices")
plt.show()

```

```python
#11
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "a": [2, 4, 6, 8, 10],
    "b": [4, 2, 4, 2, 4],
    "c": [8, 3, 7, 6, 3],
    "d": [5, 4, 4, 4, 3],
    "e": [7, 2, 7, 8, 3]
}

df = pd.DataFrame(data)

df.plot(kind='bar', x='a')
plt.xlabel("A Values")
plt.ylabel("Other Columns")
plt.title("Bar Plot from DataFrame")
plt.legend(title="Columns")
plt.show()
```

```python
#12
import matplotlib.pyplot as plt
import numpy as np

x_labels = ["A", "B", "C", "D"]
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev = [0.3314, 0.2278, 0.2836, 0.2645]

plt.bar(x_labels, mean_velocity, yerr=std_dev, capsize=5)

plt.xlabel("Categories")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars")

plt.show()
```

```python
#13
import matplotlib.pyplot as plt

x_labels = ["A", "B", "C", "D"]
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev = [0.3314, 0.2278, 0.2836, 0.2645]

plt.bar(x_labels, mean_velocity, yerr=std_dev, capsize=5)

for i in range(len(x_labels)):
    plt.text(i, mean_velocity[i], str(int(mean_velocity[i])), ha='center')

plt.xlabel("Categories")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars and Labels")

plt.show()
```

```python
#3
import matplotlib.pyplot as plt
import csv

countries = []
gold_medals = []

with open("medal.csv") as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        countries.append(row[0])
        gold_medals.append(int(row[1]))

plt.pie(gold_medals, labels=countries, autopct="%1.1f%%")

plt.title("Gold Medal Achievements in 2016 Summer Olympics")

plt.show()
```

```python
#3
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10) * 100  
y = np.random.rand(10) * 100  
sizes = np.random.rand(10) * 500  

plt.scatter(x, y, s=sizes, color="green", alpha=0.5)

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot with Different Sizes")

plt.show()
```

```python
#5
import matplotlib.pyplot as plt

group1_weight = [50, 55, 60, 65, 70]
group1_height = [150, 155, 160, 165, 170]

group2_weight = [60, 65, 70, 75, 80]
group2_height = [160, 165, 170, 175, 180]

group3_weight = [70, 75, 80, 85, 90]
group3_height = [170, 175, 180, 185, 190]

plt.scatter(group1_weight, group1_height, color="blue", label="Group 1")
plt.scatter(group2_weight, group2_height, color="green", label="Group 2")
plt.scatter(group3_weight, group3_height, color="red", label="Group 3")

plt.xlabel("Weight (kg)")
plt.ylabel("Height (cm)")
plt.title("Weight vs Height Scatter Plot")
plt.legend()

plt.show()
```
